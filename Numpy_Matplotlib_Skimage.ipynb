```python
def tileimages(img1):
    r = img1[:, :, 0]
    g = img1[:, :, 1]
    b = img1[:, :, 2]
    output = [img1,r , g, b]
    titles = ['Original', 'Red', 'Green', 'Blue']
    cmaps = [None, 'Reds', 'Greens', 'Blues']
    for i in range(4):
        plt.subplot(2, 2, i+1)
        plt.title(titles[i])
        plt.imshow(output[i], cmap=cmaps[i])
    plt.show()
```

```python
tileimages(image2)
```

```python
def quarterimage(image):
    if image.ndim ==3:
      height,width,channels = image.shape
    else:
      height,width = image.shape

    halfheight,halfwidth = height//2, width//2
                                    #tl                             tr                             bl                                br
    tl, tr, bl, br = image[:halfheight, :halfwidth], image[:halfheight, halfwidth:], image[halfheight:, :halfwidth],  image[halfheight:, halfwidth:]
    output = [tl, tr, bl, br]
    for i in range(4):
        plt.subplot(2, 2, i+1)
        plt.imshow(output[i])
    plt.show()
```

```python
def quarterimage_colors(image):
    redchannel,greenchannel,bluechannel = 0,1,2
    height,width,channels = image.shape
    halfheight,halfwidth = height//2, width//2

                                    #tl                             tr                             bl                                br
    tl, tr, bl, br = image[:halfheight, :halfwidth], image[:halfheight, halfwidth:,redchannel], image[halfheight:, :halfwidth,greenchannel],  image[halfheight:, halfwidth:,bluechannel]
    output = [tl, tr, bl, br]
    titles = ['Original', 'Red', 'Green', 'Blue']
    cmaps = [None, 'Reds', 'Greens', 'Blues']
    for i in range(4):
        plt.subplot(2, 2, i+1)
        plt.title(titles[i])
        plt.imshow(output[i], cmap=cmaps[i])
    plt.show()
```

```python
quarterimage(image2)
```

```python
import numpy as np
from skimage import data
import matplotlib.pyplot as plt

camera = data.camera()
camera[:10] = 0
mask = camera < 87
camera[mask] = 255
inds_x = np.arange(len(camera))
inds_y = (4 * inds_x) % len(camera)
camera[inds_x, inds_y] = 0

l_x, l_y = camera.shape[0], camera.shape[1]
X, Y = np.ogrid[:l_x, :l_y]
outer_disk_mask = (X - l_x / 2)**2 + (Y - l_y / 2)**2 > (l_x / 2)**2
camera[outer_disk_mask] = 0

plt.figure(figsize=(4, 4))
plt.imshow(camera, cmap='gray')
plt.axis('off')
plt.show()
```
